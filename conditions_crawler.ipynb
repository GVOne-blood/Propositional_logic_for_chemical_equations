```python
df = pd.read_excel("F:\Document\AI_SIC\Chemical_reactions_crawling\Chemical reactions datasetV2.xlsx")
print(df.head())


import pandas as pd
import re

def split_reaction(reaction):
    # Tách phản ứng tại mũi tên
    parts = re.split(r'\s*(?:→|⟶)\s*', reaction)
    if len(parts) == 2:
        # Hàm để loại bỏ số ở đầu mỗi chất và khoảng trắng
        def remove_leading_numbers_and_spaces(s):
            # Loại bỏ số ở đầu mỗi chất
            s = re.sub(r'(?:^|\+\s*)\d+(?=\s*[A-Z])', '', s)
            # Loại bỏ khoảng trắng xung quanh dấu '+'
            return re.sub(r'\s*\+\s*', '+', s.strip())

        reactants = remove_leading_numbers_and_spaces(parts[0])
        products = remove_leading_numbers_and_spaces(parts[1])
        return reactants, products
    else:
        return reaction, ''  # Trả về nguyên chuỗi nếu không tìm thấy mũi tên

# Đọc file Excel
df = pd.read_excel(r"F:\Document\AI_SIC\Chemical_reactions_crawling\Chemical reactions datasetV2.xlsx")
print(df.head())

# Áp dụng hàm cho cột chứa phản ứng
df['reactants'], df['products'] = zip(*df['Reaction'].apply(split_reaction))

# Tạo danh sách reactants và products
reactants = df['reactants'].tolist()
products = df['products'].tolist()

# In ra một vài ví dụ để kiểm tra
print("Reactants:", reactants[:5])
print("Products:", products[:5])

# Kiểm tra với ví dụ cụ thể
test_reaction = "2H2O + SO2 + 2At ⟶ H2SO4 + 2HAt"
print("Original:", test_reaction)
print("Processed:", " ⟶ ".join(split_reaction(test_reaction)))
```

# Access links by replacing url properties using requests

```python
#check for accessing links successfully
condition = []
#Catalyst, Temperature, Pressure, Other_Condition = [], [], [], []
for i in range(785, len(df)):
    url = "https://phuongtrinhhoahoc.com/phuong-trinh?chat_tham_gia=" + reactants[i] + "&chat_san_pham=" + products[i]
    response = requests.get(url)
    print('Access Successful with f{url}, status : ', response)
    soup =  BeautifulSoup(response.content, 'html.parser')
    result = (soup.find_all('p', class_ = 'font-medium'))
    for item in result:
        temp = item.text.strip().split('\n')
        condition.extend(temp)
        # print(condition)
        df.loc[i, 'Catalyst'] = condition[1] if len(condition) > 1 else 'null'
        df.loc[i, 'Temperature'] = condition[2] if len(condition) > 2 else 'null'
        df.loc[i, 'Pressure'] = condition[3] if len(condition) > 3 else 'null'
        df.loc[i, 'Other conditions'] = condition[4] if len(condition) > 4 else 'null'
    temp = []
    condition = []
        
# Lưu DataFrame đã cập nhật vào file Excel
df.to_excel("Chemical reactions datasetV2.xlsx", index=False)
```
